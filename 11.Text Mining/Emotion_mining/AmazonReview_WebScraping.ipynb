```python
def get_name(soup_obj: BeautifulSoup) -> str:
    try:
        name = soup_obj.find("span",{"class": "a-profile-name"}).get_text()
        return name
    except Exception as e:
        return 'not_found'
        print(e)
```

```python
def get_star(soup_obj: BeautifulSoup) -> str:
    try:
        star = soup_obj.find("i",{"class": "a-icon a-icon-star a-star-4 review-rating"}).get_text()
        return star
    except Exception as e:
        return 'not_found'
        print(e)
```

```python
def get_review_title(soup_obj: BeautifulSoup) -> str:
    try:
        review_title = soup_obj.find("a",{"class": "a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold"}).get_text().strip()
        return review_title
    except Exception as e:
        return 'not_found'
        print(e)
```

```python
def get_review_body(soup_obj: BeautifulSoup) -> str:
    try:
        review_body = soup_obj.find("span",{"class": "a-size-base review-text review-text-content"}).get_text().strip()
        return review_body
    except Exception as e:
        return 'not_found'
        print(e)
```

```python
def get_date(soup_obj: BeautifulSoup) -> str:
    try:
        date = soup_obj.find("span",{"class": "a-size-base a-color-secondary review-date"}).get_text().strip()
        return date
    except Exception as e:
        return 'not_found'
        print(e)
```

```python
def get_review_components(soup_obj: BeautifulSoup) -> dict:
    return {
        'User': get_name(soup_obj) , 
        'Rating': get_star(soup_obj), 
        'Title': get_review_title(soup_obj),
        'Review': get_review_body(soup_obj),
        'Date': get_date(soup_obj)
    }
```

```python
def get_soup(url):
    r = requests.get(url, headers=headers,
    params={'url': url, 'wait': 2})
    soup = BeautifulSoup(r.text, 'html.parser')
    return soup
```

```python
# Initialize list to store reviews data later on
reviewlist = []

# Function 2: look for web-tags in our soup, then append our data to reviewList
def get_reviews(soup):
    reviews = soup.find_all('div', {'data-hook': 'review'})
    try:
        for item in reviews:
            review = {
            'product': soup.title.text.replace('Amazon.in:Customer reviews: ', '').strip(),    
            'date': item.find('span', {'data-hook': 'review-date'}).text.replace('Reviewed in India on', '').strip(),
            'title': item.find('a', {'data-hook': 'review-title'}).text.strip(),
            'rating':  float(item.find('i', {'data-hook': 'review-star-rating'}).text.replace('out of 5 stars', '').strip()),
            'body': item.find('span', {'data-hook': 'review-body'}).text.strip(),
            }
            reviewlist.append(review)
    except:
        pass
```

```python
reviewlist = []

# Function 2: look for web-tags in our soup, then append our data to reviewList
def get_reviews(soup):
    reviews = soup.find_all('div', {'data-hook': 'review'})
    try:
        for item in reviews:
            review = {
            'Rating':  float(item.find('i', {'data-hook': 'review-star-rating'}).text.replace('out of 5 stars', '').strip()),
            'Title': item.find('a', {'data-hook': 'review-title'}).text.strip(),
            'Review': item.find('span', {'data-hook': 'review-body'}).text.strip(),
            'Review_Date': item.find('span', {'data-hook': 'review-date'}).text.replace('Reviewed in India on', '').strip(),
            }
            reviewlist.append(review)
    except:
        pass
```

```python
# loop through 1:x many pages, or until the css selector found only on the last page is found (when the next page button is greyed)
for x in range(1,6):
    soup = get_soup(f'https://www.amazon.in/Samsung-Storage-MediaTek-Octa-core-Processor/product-reviews/B0BMGB3CH9/ref=cm_cr_getr_d_paging_btm_next_5?ie=UTF8&reviewerType=all_reviews&pageNumber={x}')
    print(f'Getting page: {x}')
    get_reviews(soup)
    print(len(reviewlist))
```

```python
# Save results to a dataframe, then export as CSV
df = pd.DataFrame(reviewlist)
#df.to_csv(r'sony-headphones.csv', index=False)
df
```
